```python
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.io.cif import CifWriter
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer

# Define the experimental XRD pattern
exp_pattern = [(2.0, 100.0), (3.0, 50.0), (4.0, 20.0), (5.0, 10.0)]

# Create an instance of the XRDCalculator class
xrd_calculator = XRDCalculator('CuKa')

# Load the crystal structure from a CIF file
structure = Structure.from_file('structure.cif')

# Calculate the XRD pattern from the crystal structure
calc_pattern = xrd_calculator.get_pattern(structure)

# Compare the experimental and calculated XRD patterns to obtain the best match
best_match = xrd_calculator.get_best_match(exp_pattern, calc_pattern)

# Print the predicted crystal structure
predicted_structure = SpacegroupAnalyzer(structure).get_conventional_standard_structure()
CifWriter(predicted_structure, symprec=0.1).write_file('predicted_structure.cif')
```

```python
from pymatgen.analysis.diffraction.core import DiffractionPattern
from pymatgen.analysis.diffraction.xrd import get_unique_families
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer

# Define XRD parameters
wavelength = 1.54 # Angstroms
peaks = [(2.5, 25), (3.0, 40), (4.5, 60)] # List of (d, intensity) values for peaks

# Generate diffraction pattern
pattern = DiffractionPattern.from_peaks(peaks, wavelength=wavelength)

# Get unique Miller indices for diffraction peaks
families = get_unique_families(pattern)

# Generate possible structures
structures = []
for family in families:
    sg = SpacegroupAnalyzer(family[0].get_crystal_system())
    possible_structures = sg.get_symmetrized_structures(family)
    structures.extend(possible_structures)

# Filter structures based on match to diffraction pattern
matching_structures = []
for structure in structures:
    calculated_pattern = DiffractionPattern.from_structure(structure, wavelength=wavelength)
    if pattern.matches(calculated_pattern):
        matching_structures.append(structure)

# Print matching structures
for structure in matching_structures:
    print(structure)
```
